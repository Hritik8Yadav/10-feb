```python
#Ans 3: 
f=open("text.txt",'w')
f.write("I want to become a Data Scientist")
f.close()
```

```python
with open("text.txt",'r') as f:
    print(f.read())
```

```python
#Ans 4:
f=open("a.txt","w")
f.write('''open() and with open() functions are used to open the file. read the file 
write the file by replacing the words
 append the file at the end
 open the file in both read and write mode''')
f.close()

```

```python
# use of read() function- it is used to read whole content of a file
with open("a.txt",'r') as f:
    print(f.read())
    f.seek(0)
    """readline read each line after the pointer in the file"""
    print(f.readline())
    """readlines convert each line into the list"""
    print(f.readlines())
```

```python
#Ans 5: 'with' statement is used to open the file. unlike in simple open statement the we don't need to close the file
# opening by using close() statement. this make the open function more reliable.
```

```python
#Ans 6:
with open("file.txt",'w+') as f:
    f.write("this is my file.")
    f.writelines(['this ','is ','my ','file '])
```

```python
# write- it overwrite the whole file by given input.
# writelines- it adds the list into the sentence in the file.
with open('file.txt','r') as f:
    print(f.read())
```
